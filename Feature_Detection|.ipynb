```
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
```

```
os.chdir('/content/drive/My Drive/Colab Notebooks/Lane Change Detection (LCD)')
```

```
original_img = cv2.imread('sample/sample1.JPG')
plt.imshow(original_img)
plt.show()
```

```
image_2_detect = cv2.imread('sample/detect.JPG')
w,h,alpha = image_2_detect.shape[::]
plt.imshow(image_2_detect)
plt.show()
```

```
match_finder = cv2.matchTemplate(original_img,image_2_detect,cv2.TM_CCOEFF_NORMED)
plt.imshow(match_finder)
plt.show()
```

```

min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(match_finder)
top_left = max_loc
bottom_right = (top_left[0]+w, top_left[1]+h)
print(top_left,bottom_right)
```

```
clone_img =original_img.copy()
cv2.rectangle(clone_img,top_left,bottom_right,(0,0,255),10)
plt.imshow(clone_img)
plt.show()
```
